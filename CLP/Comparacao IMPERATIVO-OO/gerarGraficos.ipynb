```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_tempo_exec(tempos_execucoes: float, nomeArquivo):
    with open(nomeArquivo, 'r') as file:
        numeroExecucoes = int(file.readline())
    
        for tempo in range(numeroExecucoes):
            tempos_execucoes.append(float(file.readline()))
            
    return numeroExecucoes

nome_arquivo_c = "tempoExecucao_c.txt"
tempos_execucoes_c = []

nome_arquivo_java = "tempoExecucao_java.txt"
tempos_execucoes_java = []

numero_execucoes_c = get_tempo_exec(tempos_execucoes_c, nome_arquivo_c)
tempos_execucoes_c = [i * 1000 for i in tempos_execucoes_c]
numero_execucoes_java = get_tempo_exec(tempos_execucoes_java, nome_arquivo_java)

execucoes_c = [i for i in range(1, numero_execucoes_c + 1)]
execucoes_java = [i for i in range(1, numero_execucoes_java + 1)]
```

```python
titles = ['Tempo Execução Ordenação Imperativo - C', 'B) Tempo Execução Ordenação OO - JAVA']
ids = [0, 1]
X_values = [execucoes_c, execucoes_java]
tempos = [tempos_execucoes_c, tempos_execucoes_java]

fig, ax = plt.subplots(1, 2, figsize=(18, 8))

for id, x, y, title in zip(ids, X_values, tempos, titles):
    ax[id].plot(x, y)
    #ax[id].set_title(title)
    ax[id].set_xlabel("Quantidade de Vezes executadas")
    ax[id].set_ylabel("Tempo Gasto (ms)")
```

```python
array_c = np.array(tempos_execucoes_c)
array_java = np.array(tempos_execucoes_java)
mediaTempo_c = round(array_c.mean(), 2)
mediaTempo_java = round(array_java.mean(), 2)
print(f"Media de tempo em C {mediaTempo_c} milisegundos")
print(f"Media de tempo em java {mediaTempo_java} milisegundos")
```
